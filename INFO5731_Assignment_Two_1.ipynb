```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By

options = webdriver.ChromeOptions()


link = "https://www.imdb.com/title/tt7286456/reviews?ref_=tt_urv"
web_d = webdriver.Chrome('chromedriver',options=chrome_options)
web_d.get(link)

time.sleep(10)

mov_nam = []
date = []
content= []
page=0
while True:
    time.sleep(10)
    source = web_d.page_source
    soup = BeautifulSoup(source, 'html.parser')
    final = soup.find_all("div", {"class": "review-container"})
    if(page==3) :
      for result in final:
          mov_nam.append(result.find("a", {"class": "title"}).get_text().strip())
          date.append(result.find("span", {"review-date"}).get_text().strip())
          content.append(result.find("div", {"class": "content"}).get_text().strip())
    if len(web_d.find_elements_by_css_selector('.load-more-data')) > 0:
          web_d.find_element_by_css_selector('.load-more-data').click()
          page+=1
          if int(page)>3:
           break
    else:
          break

df = pd.DataFrame({"Title": mov_nam, "date": date, "content": content})
df.to_csv("output_review.csv")
```

# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
import pandas as pd
from nltk.corpus import stopwords
import string
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from textblob import Word


port_stem = PorterStemmer()
word_lem = WordNetLemmatizer()

pd.read_csv('output_review.csv', encoding='utf-8')
df = pd.read_csv('output_review.csv')
#data Pre-processing
df.columns = ['','Title', 'date', 'content']
df['Title'] = df['Title'].str.lower()
df['content'] = df['content'].str.lower()
df['Title']= df['Title'].str.replace(r'[^\w\s]+','')
df['content']= df['content'].str.replace(r'[^\w\s]+','')
df['Title'] = df['Title'].str.replace('\d+', '')
df['content'] = df['content'].str.replace('\d+', '')
stop = stopwords.words('english')
df['Title'] = df['Title'].apply(lambda x: [item for item in str.split(x) if item not in stop])
df['content'] = df['content'].apply(lambda x: [item for item in str.split(x) if item not in stop])
df['Title']= df['Title'].apply(lambda x : [word_lem.lemmatize(y) for y in x])
df['content']= df['content'].apply(lambda x : [word_lem.lemmatize(y) for y in x])
df['Title']= df['Title'].apply(lambda x : [port_stem.stem(y) for y in x])
df['content']= df['content'].apply(lambda x : [port_stem.stem(y) for y in x])

def punctuation(text):
  no_punc=" ".join([c for c in text if c not in string.punctuation])
  return no_punc
df['Title']= df['Title'].apply(lambda x : punctuation(x))
df['content']= df['content'].apply(lambda x : punctuation(x))

#Adding columns for clean_content and clean_title in a csv
inp_data = pd.read_csv('output_review.csv')
inp_data['clean_title'] = df['Title']
inp_data['clean_content'] = df['content']
inp_data.to_csv('output_review.csv', index=False)
```

# **Question 3**

(30 points). Write a python program to conduct **syntax and structure analysis** of the clean text you just saved above. The syntax and structure analysis includes: 

(1) Parts of Speech (POS) Tagging: Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) Constituency Parsing and Dependency Parsing: print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) Named Entity Recognition: Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
from nltk import word_tokenize, pos_tag
from spacy import displacy
from spacy.lang.en import English
from tokenize import tokenize
from collections import Counter
import nltk
import spacy

nlp = spacy.load("en_core_web_sm")


pd.read_csv('output_review.csv', encoding='utf-8')
df = pd.read_csv('output_review.csv')
df.clean_title=df.clean_title.astype(str)

tag = lambda x: pos_tag(word_tokenize(x))
df['new_title']=df['clean_title'].apply(tag)
df['new_clean_content']=df['clean_content'].apply(tag)

#Nouns
def NC(x):
   noun = []
   for (word, pos) in x:
        if pos.startswith("NN"):
            noun.append(word)
   return noun
df["noun_title"] = df["new_title"].apply(NC)
df["noun_count_title"] = df["noun_title"].str.len()
n1=df["noun_count_title"].sum()
df["noun_content"] = df["new_clean_content"].apply(NC)
df["noun_count_content"] = df["noun_content"].str.len()
n2=df["noun_count_content"].sum()

#Verbs
def VC(x):
   V = []
   for (word, pos) in x:
        if pos.startswith("VB"):
            V.append(word)
   return V
df["V_title"] = df["new_title"].apply(VC)
df["verb_title_count"] = df["V_title"].str.len()
v1=df["verb_title_count"].sum()
df["V_content"] = df["new_clean_content"].apply(VC)
df["verb_content_count"] = df["V_content"].str.len()
v2=df["verb_content_count"].sum()
#Adjectives
def AdjC(x):
   Adjs = []
   for (word, pos) in x:
        if pos.startswith("JJ"):
            Adjs.append(word)
   return Adjs
df["Adjs_title"] = df["new_title"].apply(AdjC)
df["Adj_title_count"] = df["Adjs_title"].str.len()
av1=df["Adj_title_count"].sum()
df["Adjs_content"] = df["new_clean_content"].apply(AdjC)
df["Adj_content_count"] = df["Adjs_content"].str.len()
av2=df["Adj_content_count"].sum()

#Adverbs
def AVC(x):
   Adv = []
   for (word, pos) in x:
        if pos.startswith("RB"):
            Adv.append(word)
   return Adv
df["Adv_title"] = df["new_title"].apply(AVC)
df["Adv_count_title"] = df["Adv_title"].str.len()
ad1=df["Adv_count_title"].sum()
df["Adv_content"] = df["new_clean_content"].apply(AVC)
df["Adv_count_content"] = df["Adv_content"].str.len()
ad2=df["Adv_count_content"].sum()

print("Number of nouns",n1+n2)
print("Number of verbs",v1+v2)
print("Number of Adjs",av1+av2)
print("Number of  Adv",ad1+ad2)

#Dependency parsing tree
sr=df['clean_content']
seriestosr = sr.to_string()
ner= nlp(seriestosr)
displacy.render(ner, style="dep", jupyter= True)

#named entity recognition
print([(X.text, X.label_) for X in ner.ents])
labels = [x.label_ for x in ner.ents]
Counter(labels)
```

Using Constituency Parsing, the text is broken down into smaller chunks. Non-terminals and terminals will be separated from the tree.
Nouns, verbs, adverbs, and adjectives are examples of non-terminals in the tree.

The sentence's terminals are its words.

Defintion of Dependencies:

Using the dependency parsing, words are linked together based on their connection.

Each word is represented by a vertices in the tree.

The edges of the child nodes are labeled according to their relationship to the parent node.
